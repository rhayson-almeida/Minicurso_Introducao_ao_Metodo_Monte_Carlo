# O Cálculo de $π$


```python
import numpy as np
import matplotlib.pyplot as plt

def pi(walks):
  pi=0
  pi_list=[]
  for i in range(walks):
    x = np.random.random()
    pi += 4 * np.sqrt(1 - x**2)
    pi_list.append(pi/(i+1))

  print(f"Estimativa de {pi/walks}")

  #Gráfico da estimativa de pi como função do número de walks

  plt.figure(figsize=(10,5))
  plt.plot(pi_list, color = "blue")
  plt.xlabel("Número de walks")
  plt.ylabel("Estimativa de π")
  plt.grid()

if __name__ == "__main__":
  pi(walks=100000)
```

#Melhorando a estimativa de $π$

## Uso do **teorema do limite central**

```python
def pi_medias(medias, walks):
  media = 0
  media_list=[]
  pi = 0
  for j in range(medias):
    for i in range(walks):
      x = np.random.random()
      pi += 4 * np.sqrt(1 - x**2)
    pi = pi/walks
    media += pi
    media_list.append(media/(j+1))

  #Gŕáfico da média de pi como função de número de termos

  plt.figure(figsize=(10,5))
  plt.plot(media_list, color = "blue")
  plt.xlabel("Número de termos")
  plt.ylabel("Estimativa de π")
  plt.grid()

  print(f"Estimativa de {media/medias}")


if __name__ == "__main__":
  pi_medias(medias = 1000, walks = 10000)
```
