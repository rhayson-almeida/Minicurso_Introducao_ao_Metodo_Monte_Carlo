# Visualizando o cálculo de pi com o Método Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

## A função abaixo utiliza um loop para realizar a amostragem pelo Método Monte carlo

```python
def plot_de_pi(n_pontos, tamanho_pontos):
  dentro_x, dentro_y = [], []
  fora_x, fora_y = [], []

  for _ in range(n_pontos):
    x, y = np.random.uniform(-1, 1), np.random.uniform(-1, 1)
    if x**2 + y**2 <= 1:
      dentro_x.append(x)
      dentro_y.append(y)
    else:
      fora_x.append(x)
      fora_y.append(y)

  pi_estimado = 4 * (len(dentro_x)/n_pontos)


  #Geração do plot de pi é descrito abaixo:

  plt.figure(figsize=(10, 10))

  plt.scatter(dentro_x, dentro_y, s=tamanho_pontos, c='blue', alpha=0.5)
  plt.scatter(fora_x, fora_y, s=tamanho_pontos, c='red', alpha=0.5)

  circunferencia = plt.Circle((0,0), 1, color='black', fill=False, lw=2)
  plt.gca().add_patch(circunferencia)

  plt.xlim(-1.1, 1.1)
  plt.ylim(-1.1, 1.1)
  plt.gca().set_aspect('equal')
  plt.title(f"Estimativa de $\pi \approx$ {pi_estimado:.4f} \n Pontos: {n_pontos}", fontsize=15)
  plt.legend(markerscale=1.5) # Aumenta o ponto também na legenda
  plt.title(f"Estimativa de pi: {pi_estimado:.4f}\nPontos: {n_pontos}", fontsize=15)
```

```python
plot_de_pi(100,tamanho_pontos=150)
```
