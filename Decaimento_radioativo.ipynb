# **Decaimento Radioativo**

## Um exemplo de aleatoriedade de um sistema real simulado com o método Monte Carlo



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definição dos parâmetros do decaimento

N0 = 10000                          # Número inicial de núcleos
lamb = 0.5                          # Parâmetro de decaimento
passos_totais = 10                  # Número de passos no Método Monte Carlo
tempo = np.arange(passos_totais)    # Intervalo de tempo do decaimento
```

```python
def decaimento(N0, lamb, passo_tempo, dt=1.0):
  probabilidade_de_decaimento = lamb * dt
  N_t = np.zeros(passos_totais)
  N_t[0] = N0

  nucleos_atuais = N0

  for t in range(1, passo_tempo):
    if nucleos_atuais <= 0:
      N_t[t] = 0
      continue

    numero_aleatorio  = np.random.rand(nucleos_atuais)
    nucleos_decaidos = np.sum(numero_aleatorio < probabilidade_de_decaimento)
    nucleos_atuais -= nucleos_decaidos
    N_t[t] = nucleos_atuais

  return N_t
```

```python
simulacao = decaimento(N0, lamb, passos_totais)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tempo, simulacao, label="Simulação Monte Carlo", color="blue", linewidth=1.5)

plt.xlabel("Tempo", fontsize=12)
plt.ylabel("Número de Núcleos Restantes", fontsize=12)
plt.title(f"Decaimento Radioativo via Monte Carlo ($N_0={N0}$, $\lambda={lamb}$)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.5)
plt.show()
```
